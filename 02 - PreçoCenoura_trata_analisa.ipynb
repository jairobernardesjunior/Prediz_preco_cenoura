```python
# importa bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# pd.set_option('display.max_rows', None)
```

```python
# carrega arquivos limpos e semi-tratados
isx = pd.read_csv('arquivos_tratados/isx.csv')
iax = pd.read_csv('arquivos_tratados/iax.csv')
iaaax = pd.read_csv('arquivos_tratados/iaaax.csv')
igx = pd.read_csv('arquivos_tratados/igx.csv')

vsx = pd.read_csv('arquivos_tratados/vsx.csv')
vax = pd.read_csv('arquivos_tratados/vax.csv')
vaaax = pd.read_csv('arquivos_tratados/vaaax.csv')
vgx = pd.read_csv('arquivos_tratados/vgx.csv')
```

```python
# verifica a distribuiçao dos tipos de cenouras inverno A e verão AAA
fig,ax = plt.subplots(1,6,figsize=(10,5))

sns.histplot(iax.Preço_kg, bins=20, kde=True, ax=ax[0])
sns.histplot(iax.temp_max, bins=20, kde=True, ax=ax[1])
sns.histplot(iax.precipitação_mm, bins=20, kde=True, ax=ax[2])
ax[0].set_title('Inverno A')
ax[1].set_title('Inverno A')
ax[2].set_title('Inverno A')

sns.histplot(vaaax.Preço_kg, bins=20, kde=True, ax=ax[3])
sns.histplot(vaaax.temp_max, bins=20, kde=True, ax=ax[4])
sns.histplot(vaaax.precipitação_mm, bins=20, kde=True, ax=ax[5])
ax[3].set_title('Verão AAA')
ax[4].set_title('Verão AAA')
ax[5].set_title('Verão AAA')

plt.show()
```

- Como mencionado anteriormente existe assimetria mais acentuada na variável precipitação nas duas distribuições tanto na cenoura de verão quanto na cenoura de inverno. Mais adiante os outliers serão removidos e será feita uma normalização para tentar atenuar essa assimetria.

```python
# verifica outliers de preço e temp_max para cenoura inverno A e cenoura verão AAA
fig, axs = plt.subplots(2, 3, figsize=(10,5))

# linha 0 coluna 0
axs[0, 0].boxplot(iax.Preço_kg)
axs[0, 0].set_title('Preço em kg - inverno A')

# linha 0 coluna 1
axs[0, 1].boxplot(iax.temp_max, 1)
axs[0, 1].set_title('Temp max - inverno A')

# linha 0 coluna 2
axs[0, 2].boxplot(iax.precipitação_mm, 1)
axs[0, 2].set_title('Precipitação_mm - inverno A')

# linha 1 coluna 0
axs[1, 0].boxplot(vaaax.Preço_kg)
axs[1, 0].set_title('Preço em kg - verão AAA')

# linha 1 coluna 1
axs[1, 1].boxplot(vaaax.temp_max, 1)
axs[1, 1].set_title('Temp max - verão AAA')

# linha 1 coluna 2
axs[1, 2].boxplot(vaaax.precipitação_mm, 1)
axs[1, 2].set_title('Precipitação - verão A')

plt.subplots_adjust(left=0.95, bottom=0.1, right=2, top=0.9, wspace=0.8, hspace=0.4)

plt.show()
```

```python
# retira e verifica outliers de preço e temp_max
vaaax = vaaax[(vaaax.Preço_kg > 2.6) & (vaaax.Preço_kg < 6.5)]

fig, axs = plt.subplots(2, 3, figsize=(10,5))

# linha 0 coluna 0
axs[0, 0].boxplot(iax.Preço_kg)
axs[0, 0].set_title('Preço em kg - inverno A')

# linha 0 coluna 1
axs[0, 1].boxplot(iax.temp_max, 1)
axs[0, 1].set_title('Temp max - inverno A')

# linha 0 coluna 2
axs[0, 2].boxplot(iax.precipitação_mm, 1)
axs[0, 2].set_title('Precipitação_mm - inverno A')

# linha 1 coluna 0
axs[1, 0].boxplot(vaaax.Preço_kg)
axs[1, 0].set_title('Preço em kg - verão AAA')

# linha 1 coluna 1
axs[1, 1].boxplot(vaaax.temp_max, 1)
axs[1, 1].set_title('Temp max - verão AAA')

# linha 1 coluna 2
axs[1, 2].boxplot(vaaax.precipitação_mm, 1)
axs[1, 2].set_title('Precipitação - verão A')

plt.subplots_adjust(left=0.95, bottom=0.1, right=2, top=0.9, wspace=0.8, hspace=0.4)

plt.show()
```

```python
# faz a normalização das variáveis para melhorar a simetria
iaxN = pd.DataFrame()
#iaxN['Preço_kg'] = (iax['Preço_kg'] - iax['Preço_kg'].min()) / (iax['Preço_kg'].max() - iax['Preço_kg'].min())

iaxN['temp_max'] = (iax['temp_max'] - iax['temp_max'].min()) / (iax['temp_max'].max() - iax['temp_max'].min())
iaxN['precipitação_mm'] = (iax['precipitação_mm'] - iax['precipitação_mm'].min()) / \
                          (iax['precipitação_mm'].max() - iax['precipitação_mm'].min())
iaxN['Preço_kg_orig'] = iax['Preço_kg']
iaxN['temp_max_orig'] = iax['temp_max']
iaxN['precipitação_mm_orig'] = iax['precipitação_mm']

vaaaxN = vaaax.copy()
#vaaaxN['Preço_kg'] = (vaaax['Preço_kg'] - vaaax['Preço_kg'].min()) / (vaaax['Preço_kg'].max() - vaaax['Preço_kg'].min())

vaaaxN['temp_max'] = (vaaax['temp_max'] - vaaax['temp_max'].min()) / (vaaax['temp_max'].max() - vaaax['temp_max'].min())
vaaaxN['precipitação_mm'] = (vaaax['precipitação_mm'] - vaaax['precipitação_mm'].min()) / \
                          (vaaax['precipitação_mm'].max() - vaaax['precipitação_mm'].min())
vaaaxN['Preço_kg_orig'] = vaaax['Preço_kg']
vaaaxN['temp_max_orig'] = vaaax['temp_max']
vaaaxN['precipitação_mm_orig'] = vaaax['precipitação_mm']

iaxN['Preço_kg'] = iax['Preço_kg']
iaxN['data_med'] = iax['data_med']
vaaaxN['Preço_kg'] = vaaax['Preço_kg']
vaaaxN['data_med'] = vaaax['data_med']

iax = iaxN
vaaax = vaaaxN

print('iax')
print(iaxN.head(3))

print(' ')
print('vaaax')
print(vaaaxN.head(3))
```

```python
# verifica a distribuiçao dos tipos de cenouras inverno A e verão AAA depois da normalização
fig,ax = plt.subplots(1,6,figsize=(10,5))

sns.histplot(iax.Preço_kg, bins=20, kde=True, ax=ax[0])
sns.histplot(iax.temp_max, bins=20, kde=True, ax=ax[1])
sns.histplot(iax.precipitação_mm, bins=20, kde=True, ax=ax[2])
ax[0].set_title('Inverno A')
ax[1].set_title('Inverno A')
ax[2].set_title('Inverno A')

sns.histplot(vaaax.Preço_kg, bins=20, kde=True, ax=ax[3])
sns.histplot(vaaax.temp_max, bins=20, kde=True, ax=ax[4])
sns.histplot(vaaax.precipitação_mm, bins=20, kde=True, ax=ax[5])
ax[3].set_title('Verão AAA')
ax[4].set_title('Verão AAA')
ax[5].set_title('Verão AAA')

plt.show()
```

- Após retirada dos outliers e a normalização houve uma pequena redução na assimetria, porém, na distribuição da precipitação continua acentuada.

Nota-se que a temperatura manteve-se elevada durante os anos 2022 a 2024 com picos acima de 36°C. Os preços da cenoura de inverno A tem uma alta expressiva em janeiro-2023 com chuvas intensas no período chuvoso que se apresenta desde dezembro-2022, ou seja há mais de 2 meses já estava chovendo intensamente, a temperatura já se apresentava alta e se mantém. A cenoura de verão começa a ter um preço crescente a partir de fevereiro-2023 coincidindo com chuvas intensas, porém em menor escala, nesse período oscilando para mais e para menos nesse patamar mais alto até agosto-2023, a temperatura se mantém alta para esse período em função do fenômeno do el niño. Observa-se uma interrupção na alta de preços da cenoura de inverno A no início de janeiro-2023 ficando constante provavelmente pelo fim da colheita da safra de inverno, como também se verifica uma interrupção dos dados em janeiro-2024 encerrando a colheita da safra de inverno também. Os preços da cenoura de inverno cai em dezembro-2023, mais precisamente entre novembro-dezembro-2023. O preço da cenoura de verão, linha vermelha, apresenta o melhor desempenho com altos preços, tendo uma alta significativa a partir de janeiro-2024, mantendo preço elevado até março-2024.
