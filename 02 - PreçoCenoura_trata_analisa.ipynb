```python
# calcula e mostra o coeficiente de variação (desvio padrão / média)
def calcula_mostra_cv(xdescribe, xtipo):
    print(' ')
    print(xtipo)
    for i in [0, 1, 2]:
        print('cv', xdescribe.columns[i], ' =', round(xdescribe.iloc[7, i]/xdescribe.iloc[1, i], 4))     
```

```python
# analisa o coeficiente de variação das variáveis de cada tipo de cenoura
calcula_mostra_cv(isx.describe(), 'isx')
calcula_mostra_cv(iax.describe(), 'iax')
calcula_mostra_cv(iaaax.describe(), 'iaaax')
calcula_mostra_cv(igx.describe(), 'igx')

calcula_mostra_cv(vsx.describe(), 'vsx')
calcula_mostra_cv(vax.describe(), 'vax')
calcula_mostra_cv(vaaax.describe(), 'vaaax')
calcula_mostra_cv(vgx.describe(), 'vgx')
```

```python
# verifica outliers de preço e temp_max após retirada dos outliers
iax = iax[(iax.Preço_kg < 2.4)]
vaaax = vaaax[(vaaax.Preço_kg > 2.6) & (vaaax.Preço_kg < 6.5)]

fig, axs = plt.subplots(2, 3, figsize=(10,5))

# linha 0 coluna 0
axs[0, 0].boxplot(iax.Preço_kg)
axs[0, 0].set_title('Preço em kg - inverno A')

# linha 0 coluna 1
axs[0, 1].boxplot(iax.temp_max, 1)
axs[0, 1].set_title('Temp max - inverno A')

# linha 0 coluna 2
axs[0, 2].boxplot(iax.precipitação_mm, 1)
axs[0, 2].set_title('Precipitação_mm - inverno A')

# linha 1 coluna 0
axs[1, 0].boxplot(vaaax.Preço_kg)
axs[1, 0].set_title('Preço em kg - verão AAA')

# linha 1 coluna 1
axs[1, 1].boxplot(vaaax.temp_max, 1)
axs[1, 1].set_title('Temp max - verão AAA')

# linha 1 coluna 2
axs[1, 2].boxplot(vaaax.precipitação_mm, 1)
axs[1, 2].set_title('Precipitação - verão A')

plt.subplots_adjust(left=0.95, bottom=0.1, right=2, top=0.9, wspace=0.8, hspace=0.4)

plt.show()
```

```python
# normalizando variáveis
iaxN = pd.DataFrame()
#iaxN['Preço_kg'] = (iax['Preço_kg'] - iax['Preço_kg'].min()) / (iax['Preço_kg'].max() - iax['Preço_kg'].min())

iaxN['temp_max'] = (iax['temp_max'] - iax['temp_max'].min()) / (iax['temp_max'].max() - iax['temp_max'].min())
iaxN['precipitação_mm'] = (iax['precipitação_mm'] - iax['precipitação_mm'].min()) / \
                          (iax['precipitação_mm'].max() - iax['precipitação_mm'].min())
iaxN['Preço_kg_orig'] = iax['Preço_kg']
iaxN['temp_max_orig'] = iax['temp_max']
iaxN['precipitação_mm_orig'] = iax['precipitação_mm']

vaaaxN = vaaax.copy()
#vaaaxN['Preço_kg'] = (vaaax['Preço_kg'] - vaaax['Preço_kg'].min()) / (vaaax['Preço_kg'].max() - vaaax['Preço_kg'].min())

vaaaxN['temp_max'] = (vaaax['temp_max'] - vaaax['temp_max'].min()) / (vaaax['temp_max'].max() - vaaax['temp_max'].min())
vaaaxN['precipitação_mm'] = (vaaax['precipitação_mm'] - vaaax['precipitação_mm'].min()) / \
                          (vaaax['precipitação_mm'].max() - vaaax['precipitação_mm'].min())
vaaaxN['Preço_kg_orig'] = vaaax['Preço_kg']
vaaaxN['temp_max_orig'] = vaaax['temp_max']
vaaaxN['precipitação_mm_orig'] = vaaax['precipitação_mm']

iaxN['Preço_kg'] = iax['Preço_kg']
iaxN['data_med'] = iax['data_med']
vaaaxN['Preço_kg'] = vaaax['Preço_kg']
vaaaxN['data_med'] = vaaax['data_med']

iax = iaxN
vaaax = vaaaxN

print('iax')
print(iaxN.head(3))

print(' ')
print('vaaax')
print(vaaaxN.head(3))
```
