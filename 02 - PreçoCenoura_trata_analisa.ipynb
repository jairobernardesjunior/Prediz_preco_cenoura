```python
# calcula e mostra o coeficiente de variação (desvio padrão / média)
def calcula_mostra_cv(xdescribe, xtipo):
    print(' ')
    print(xtipo)
    for i in [0, 1, 2]:
        print('cv', xdescribe.columns[i], ' =', round(xdescribe.iloc[7, i]/xdescribe.iloc[1, i], 4))     
```

```python
# analisa o coeficiente de variação das variáveis de cada tipo de cenoura
calcula_mostra_cv(isx.describe(), 'isx')
calcula_mostra_cv(iax.describe(), 'iax')
calcula_mostra_cv(iaaax.describe(), 'iaaax')
calcula_mostra_cv(igx.describe(), 'igx')

calcula_mostra_cv(vsx.describe(), 'vsx')
calcula_mostra_cv(vax.describe(), 'vax')
calcula_mostra_cv(vaaax.describe(), 'vaaax')
calcula_mostra_cv(vgx.describe(), 'vgx')
```

```python
# retira os outliers do Preço_kg para visualizar apenas e verificar se fica igual quanto retirar do dataframe
fig,ax = plt.subplots(1,2,figsize=(20,2))

sns. boxplot (x='Preço_kg', data=iax, ax=ax[0], showfliers = False)
sns. boxplot (x='Preço_kg', data=vaaax, ax=ax[1], showfliers = False)
ax[0].set_title('Preço_kg inverno A')
ax[1].set_title('Preço_kg verão AAA')

plt.show()
```

```python
# retira outliers do Preço_kg nos data frames iax e vaaax e verifica se ficou igual à vizualização anterior
iax = iax[(iax.Preço_kg < 2.4)]
vaaax = vaaax[(vaaax.Preço_kg > 2.6) & (vaaax.Preço_kg < 6.5)]

fig,ax = plt.subplots(1,2,figsize=(20,2))

sns. boxplot (x='Preço_kg', data=iax, ax=ax[0], showfliers = True)
sns. boxplot (x='Preço_kg', data=vaaax, ax=ax[1], showfliers = True)
ax[0].set_title('Preço_kg inverno A')
ax[1].set_title('Preço_kg verão AAA')

plt.show()
```

```python
# verifica a distribuição depois de retirar os outliers
fig,ax = plt.subplots(1,2,figsize=(10,5))

sns.histplot(iax.Preço_kg, bins=20, kde=True, ax=ax[0])
sns.histplot(vaaax.Preço_kg, bins=20, kde=True, ax=ax[1])
ax[0].set_title('Inverno A')
ax[1].set_title('Verão AAA')

plt.show()
```
